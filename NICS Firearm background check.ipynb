```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./dataset/nics-firearm-background-checks.csv")
```

```python
df.groupby(['month', 'state'])["permit"].apply(lambda x : x.astype(int).sum())
```

```python
df.head(100)
```

```python
ax=df.plot(x="state", y="permit_recheck", kind="bar")
df.plot(x="state",y="handgun", kind="bar", ax=ax, color="blue")
df.plot(x="state", y="long_gun", kind="bar", ax=ax, color="red")
plt.show()
```

```python
ax=df.plot(x="permit", y="permit_recheck", kind="bar")
df.plot(x="permit",y="handgun", kind="bar", color="blue")
plt.show()
```
