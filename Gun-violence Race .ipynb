```python
#chose the following variable that we will be focasing on 

df1= df[['year', 'intent','police','age','race','place','police']]
df1
```

```python
#Getting all the count of each type of intent per race per year
# It will filter first per year then per race then finaly count all the different intent realated to those filter

sumInci=df1.groupby(['year','race','intent'])["age"].apply(lambda x : x.astype(int).count()).to_frame('total_incident_count').reset_index()
sumInci
```
